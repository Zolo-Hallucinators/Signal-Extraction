## Main URL NEWS API

```python
import requests
# company = "rambus"
company = "palantir"
url = (f'https://newsapi.org/v2/everything?q={company}&language=en&searchIn=title&from=2025-09-01&to=2025-09-08&sortBy=popularity&page=1&apiKey=c494f280427646c78473013990b3cd45')
response = requests.get(url)
print(response.json())

# Stats (2025-09-01 to 2025-09-08):
# Rambus: title: 3 | description: 3 | content: 4
# Palantir: title: 78 | description: 72 | content: 155
```

```python
import json

# Convert response.json() (dict)  JSON string
response_str = json.dumps(response.json(), indent=2)

# Load back into dict
data = json.loads(response_str)

# Extract URLs
urls = [article['url'] for article in data.get('articles', [])]

# Print URLs
for url in urls:
    print(url)
```

## Beautiful Soup

```python
from bs4 import BeautifulSoup

articles_data = []

for link in urls:
    print(f"Scraping: {link}")
    try:
        res = requests.get(link, timeout=10)
        soup = BeautifulSoup(res.text, 'html.parser')

        # Extract title
        title = soup.title.string.strip() if soup.title else ""

        # Extract all paragraph text
        paragraphs = [p.get_text(strip=True) for p in soup.find_all('p')]
        content = " ".join(paragraphs)

        articles_data.append({
            "url": link,
            "title": title,
            "content": content,
            "content_size": len(content)   # number of characters in content
        })

    except Exception as e:
        articles_data.append({
            "url": link,
            "title": "",
            "content": f"Error scraping: {e}",
            "content_size": 0
        })

for article in articles_data:
    print(f"{article['url']}: {article['content_size']}")
```

```python
# Count how many have content_size = 0
zero_count = sum(1 for article in articles_data if article['content_size'] == 0)
total_articles = len(articles_data)

# Total sum of content_size
total_content_size = sum(article['content_size'] for article in articles_data)

print("Articles with content_size = 0:", zero_count)
print("Total content_size across all articles:", total_content_size)
print("Total articles scraped:", total_articles)
```

## Scrapy Library

```python
import scrapy
from scrapy.crawler import CrawlerProcess

# urls = [
#     "https://example.com",
#     "https://www.wikipedia.org"
# ]

class ArticleSpider(scrapy.Spider):
    name = "article_spider"
    start_urls = urls

    def parse(self, response):
        title = response.css("title::text").get(default="").strip()
        content = " ".join(p.get().strip() for p in response.css("p::text"))
        yield {
            "url": response.url,
            "title": title,
            "content": content,
            "content_size": len(content),
        }

process = CrawlerProcess(settings={
    "LOG_LEVEL": "ERROR",   # keep output clean
})
results = []
process.crawl(ArticleSpider)
process.start()

# Collect results from items
for spider in process.crawlers:
    results.extend(spider.spider.crawler.stats.get_stats().get("item_scraped_count", []))

print(results)
```

## Selenium

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time

# List of article URLs
# urls = [
#     "https://finance.yahoo.com/news/evercore-picks-astera-macom-top-150433255.html",
#     "https://www.etfdailynews.com/2025/09/07/trexquant-investment-lp-sells-75897-shares-of-rambus-inc-rmbs/",
#     "https://www.etfdailynews.com/2025/09/05/rambus-inc-rmbs-shares-sold-by-alphaquest-llc/",
#     "https://biztoc.com/x/81a6ba9d5a3bd19c"
# ]

# Setup Chrome options for headless browsing
chrome_options = Options()
chrome_options.add_argument("--headless")  # run in background
chrome_options.add_argument("--disable-gpu")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--window-size=1920,1080")

# Initialize driver
service = Service("chromedriver.exe")  # path to your chromedriver
driver = webdriver.Chrome(service=service, options=chrome_options)

# Store scraped articles
articles_data = []

for link in urls:
    try:
        driver.get(link)
        time.sleep(3)  # wait for JS to render

        # Get title
        title = driver.title

        # Get all paragraph texts
        paragraphs = driver.find_elements(By.TAG_NAME, "p")
        content = " ".join([p.text for p in paragraphs])

        articles_data.append({
            "url": link,
            "title": title,
            "content": content,
            "content_size": len(content)
        })

    except Exception as e:
        articles_data.append({
            "url": link,
            "title": "",
            "content": f"Error scraping: {e}",
            "content_size": 0
        })

driver.quit()
```

## Playwright

```python
import asyncio
from playwright.async_api import async_playwright

# urls = [
#     "https://finance.yahoo.com/news/evercore-picks-astera-macom-top-150433255.html",
#     "https://www.etfdailynews.com/2025/09/07/trexquant-investment-lp-sells-75897-shares-of-rambus-inc-rmbs/",
#     "https://www.etfdailynews.com/2025/09/05/rambus-inc-rmbs-shares-sold-by-alphaquest-llc/",
#     "https://biztoc.com/x/81a6ba9d5a3bd19c"
# ]

articles_data = []

async def scrape_articles():
    async with async_playwright() as p:
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()

        for link in urls:
            try:
                await page.goto(link, timeout=15000)
                await page.wait_for_load_state("domcontentloaded")

                # Get title
                title = await page.title()

                # Get all paragraphs
                paragraphs = await page.query_selector_all("p")
                content = " ".join([await p.inner_text() for p in paragraphs])

                articles_data.append({
                    "url": link,
                    "title": title,
                    "content": content,
                    "content_size": len(content)
                })

            except Exception as e:
                articles_data.append({
                    "url": link,
                    "title": "",
                    "content": f"Error scraping: {e}",
                    "content_size": 0
                })

        await browser.close()

# Run the async function in Jupyter
await scrape_articles()

# Print URL and content_size
for article in articles_data:
    print(f"{article['url']}: {article['content_size']}")
```
